```python
df["date_added"] = pd.to_datetime(df['date_added'])            #Changing date pattern i.e. September 8,2017 to 2017-09-08
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

```python
df.head(10)
```

```python
#Data is investigated here for total no. of shows, column names, dataset variables, data-type of each column.
def data_inv(df):     
    print('netflix movies and shows: ',df.shape[0])
    print('dataset variables: ',df.shape[1])
    print('-'*10)
    print('dateset columns: \n')
    print(df.columns)
    print('-'*10)
    print('data-type of each column: \n')
    print(df.dtypes)
    print('-'*10)
    print('missing rows in each column: \n')
    c=df.isnull().sum()
    print(c[c>0])
data_inv(df)
```

```python
dups=df.duplicated(['title','country','type','release_year'])     #It creates duplicates
df[dups]
```

```python
df=df.drop_duplicates(['title','country','type','release_year'])       #It drops dulicates
```

```python
df=df.drop('show_id',axis=1)
```

```python
df['cast']=df['cast'].replace(np.nan,'Unknown')                        #It replaces NaN with Unknown
def cast_counter(cast):
    if cast=='Unknown':
        return 0
    else:
        lst=cast.split(', ')
        length=len(lst)
        return length
df['number_of_cast']=df['cast'].apply(cast_counter)
df['cast']=df['cast'].replace('Unknown',np.nan)
```

```python
sns.countplot(netflix['type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type: MOVIE AND T.V SHOWS')
```

```python
sns.countplot(netflix['rating'])
sns.countplot(netflix['rating']).set_xticklabels(sns.countplot(netflix['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('RATING OF SHOWS V/S MOVIES')
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('RELATION B.W MOVIE AND T.V SHOWS')
plt.show()
```

```python
labels = ['Movies', 'TV shows']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type (Movie/TV shows)', fontsize = 25)
plt.legend()
plt.show()
```

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.title('PIE CHART FOR T.V RATINGS FOR MOVIES')
plt.show()
```

```python
top_productive_countries=df[(df['country']=='United States')|(df['country']=='India')|(df['country']=='United Kingdom')|(df['country']=='Japan')|
                             (df['country']=='Canada')|(df['country']=='Spain')]
plt.figure(figsize=(10,8))
sns.countplot(x='country',hue='type',data=top_productive_countries)
plt.title('COMPARISON B.W TYPES (movies/shows) ACCORDNG TO COUNTRIES PRODUCTION')
plt.show()
```
